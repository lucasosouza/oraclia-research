```python
%load_ext autoreload
%autoreload 2

import RNN_data_transform as dt 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start_date_train='2002-01-01'
end_date_train='2015-12-31'
start_date_test = '2016-01-01'
end_date_test='2016-06-30'

strategies = {
    'Conservative': (10, .09) 
}

stocks = ['ITUB4', 'BBDC4', 'ABEV3', 'PETR4', 'VALE5', 'VALE3', 'BBAS3', 'PETR3']

indices = []
for strategy in strategies.keys():
    for symbol in stocks:
        indices.append((strategy, symbol))

# Download basic data
market_df = dt.download_market_data(start_date=start_date_train, end_date=end_date_test)

for strategy, (span, profit) in strategies.items():
    # initialize data structures
    data = {}
    # train classifiers
    for symbol in stocks:
        ##### Get Data ##################################
        print('Downloading training data for {}'.format(symbol))
        data[symbol + '_X'], data[symbol + '_y'], column_names = \
            dt.gen_data(symbol, market_df, start_date_train, end_date_train, span=span, profit=profit)
```
